# 3.What is the Correlation between the features?

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)
print("Correlation between features:")
print(corr_matrix)
```

# 5.Distribution of Ratings?

  # 7.Finding the best Month for releasing content?

```python
df.head(5)
```

```python
df[df.date_added.isnull()]
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors='coerce')
```

```python
df[df.date_added.isnull()]
```

```python
df["month"] = df["date_added"].dt.month_name()
month_count = df["month"].value_counts()
print("Number of releases per month:")
print(month_count)

# Best month
best_month = month_count.idxmax()
print("\nBest month for releasing content:", best_month)
```

```python
df.isnull().sum()
```

```python
df["month"].value_counts().plot(kind='bar', figsize=(5,5))
plt.title("Content Releases by Month on Netflix")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```

# 8.Highest watched genres on Netflix?

```python
df.listed_in.value_counts()
```

```python
# Split the 'listed_in' column because each entry may contain multiple genres
genres = df['listed_in'].str.split(', ', expand=True).stack()
# Count genre frequency
genre_count = genres.value_counts()
print("Most popular genres on Netflix:")
print(genre_count)
```

```python
genre_count.head(10).plot(kind='bar', figsize=(5,5))
plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

# 9.Released movie over the years?

```python
movies = df[df['type'] == 'Movie']
movies_per_year = movies['release_year'].value_counts().sort_index()
movies_per_year
```

```python
movies_per_year.plot(kind='line', figsize=(5,5))
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Movies Released Over the Years on Netflix')
plt.show()
```

# 14.Show top 10 director, who gave the highest number of TV shows & Movies to Netflix?

```python
# Step 1: Remove rows with missing directors
df_directors = df[df['director'].notna()]
# Step 2: Count number of titles per director
top_directors = df_directors['director'].value_counts().head(10)
# Step 3: Display the result
print(top_directors)
```

# 15.In how many movies/ tv shows, 'tom Cruise' was cast?

```python
tom_cruise_titles = df[df['cast'].str.contains('Tom Cruise', case=False, na=False)][['title', 'type']]
print(tom_cruise_titles)
```
